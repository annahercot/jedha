```python
import urllib
import bs4
```

```python
url = "https://www.imdb.com/chart/top"

from urllib import request
request_text = request.urlopen(url).read()
print(request_text[:1000])
```

```python
page = bs4.BeautifulSoup(request_text, 'lxml')
print(page)
```

```python
base_url = "https://www.imdb.com"
imdb = []
for item in page.find('tbody', {'class':'lister-list'}).findAll('tr')[1:20]:
    title = item.find('td', {'class': 'titleColumn'}).find('a').text
    href = item.find({'a'}).get('href')
    rating = item.find('td', {'class': 'ratingColumn'}).text
    movie_url = (base_url + href)
    movie_text = request.urlopen(movie_url).read()
    movie_page = bs4.BeautifulSoup(movie_text, 'lxml')
    release = movie_page.find('span', {'id':'titleYear'}).find('a').text
    votes = movie_page.find('div', {'class':'imdbRating'}).find('a').text
    actors = [actor.select('td')[1].text.strip() for actor 
              in movie_page.find('table', {'class':'cast_list'}).findAll('tr', {'class':['odd', 'even']})]
    movie = {
        "title": title,
        "release_date": release,
        "actors": actors,
        "rating": rating,
        "number_of_votes": votes,
        "link":movie_url
    }
    imdb.append(movie)
        
```

```python
print(imdb)
```
