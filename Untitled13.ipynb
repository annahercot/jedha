```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
n = 50000
array = np.random.exponential(scale=5, size=n)
array
```

```python
sns.distplot(array)
```

```python
split_arrays = np.hsplit(array, 1000)
means = np.mean(split_arrays, axis=1)
```

```python
sns.distplot(means)
```

```python
std = np.std(means)
std
```

```python
mean = np.mean(means)
mean
```

```python
normal = np.random.normal(loc = mean, scale = std, size = 1000)
```

```python
sns.distplot(normal)
sns.distplot(means)
```

```python
nn = 5000
arr = np.random.binomial(1, 0.7, size=nn)
arr
```

```python
sns.distplot(arr)
```

```python
split_arr = np.hsplit(arr, 1000)
mid = np.mean(split_arr, axis=1)
```

```python
sns.distplot(mid)
```

```python
mstd = np.std(mid)
mmid = np.mean(mid)
norm = np.random.normal(loc = mmid, scale = mstd, size = 1000)
```

```python
sns.distplot(norm)
sns.distplot(mid)
```
