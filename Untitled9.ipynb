```python
dataset = pd.read_csv("/users/vincentsalamand/Downloads/train.csv")
del dataset["PassengerId"], dataset["Ticket"], dataset["Embarked"], dataset["Fare"]

dataset = dataset.interpolate()
dataset = dataset.fillna("Unknown")

dataset["Cabin_group"] = dataset.Cabin.apply(lambda x: x[0])
del dataset["Cabin"]

dataset["Family"] = dataset.SibSp + dataset.Parch
del dataset["SibSp"], dataset["Parch"]

dataset["Family_name"] = dataset.Name.str.len()

del dataset["Name"]


dataset = pd.get_dummies(dataset, drop_first = True)

dataset.head()
```

```python
y = dataset.Survived
X = dataset.loc[:, dataset.columns!="Survived"]
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x.fit(X)
X = sc_x.transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.25,
                                                   random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```
