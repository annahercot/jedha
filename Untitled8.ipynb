```python
import pandas as pd
import numpy as np
```

```python
house_price = pd.read_csv("/users/vincentsalamand/Downloads/house_price.csv")
nombre_de_sdb = pd.read_csv("/users/vincentsalamand/Downloads/nombre_de_sdb.csv")
nombre_de_chambres = pd.read_csv("/users/vincentsalamand/Downloads/nombre_de_chambres.csv")
superficie = pd.read_csv("/users/vincentsalamand/Downloads/superficie.csv", header = None)
```

```python
nombre_de_chambres = nombre_de_chambres.rename(columns={"0":"id", "1":"nombre_de_chambres"})
nombre_de_sdb = nombre_de_sdb.rename(columns={"0":"id", "2":"nombre_de_sdb"})
superficie = superficie.rename(columns={0:"id", 1:"superficie"})
```

```python
superficie.head()
```

```python
pd.concat([superficie, nombre_de_chambres, nombre_de_sdb, house_price], axis = 1).head()
```

```python
dataset = superficie.merge(nombre_de_sdb, on="id").merge(nombre_de_chambres, on="id").merge(house_price, on="id")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset = pd.concat([dataset.iloc[:, -2], dataset.iloc[:,-1]], axis =1)
dataset.head()
```

```python
dataset.house_price.mean()
```

```python
dataset.groupby("nombre_de_chambres").mean().astype(int)
```

```python
dataset[(dataset.house_price > dataset.house_price.quantile(0.1)) & (dataset.house_price < dataset.house_price.quantile(0.9))].house_price.describe()
```

```python

dataset["size"] = dataset.superficie.apply(lambda x: "XL" if x > 25000 
                                             else "L" if x > 20000  
                                             else "M" if x > 15000 
                                             else "S")
dataset
```
